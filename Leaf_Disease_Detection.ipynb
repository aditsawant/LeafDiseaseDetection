```python
train_images = np.load("./AlexNet_PlantPathology/all_train_images.npy")
test_images = np.load("./AlexNet_PlantPathology/all_test_images.npy")
```

```python
def AlexNet(input_shape = (227,227,3), classes = 4):
    #Input
    X_input = Input(input_shape)
    
    #Layer1
    X = Conv2D(96, kernel_size= (11,11), strides = (4,4), padding= 'valid', kernel_initializer= glorot_uniform(seed= 0))(X_input)
    X = BatchNormalization(axis=3)(X)
    X = Activation('relu')(X)
    #Applying MaxPooling
    X = MaxPooling2D(pool_size= (3,3), strides= (2,2))(X)
    
    #Layer2
    X = Conv2D(256, kernel_size= (5,5), padding= 'same', kernel_initializer= glorot_uniform(seed= 0))(X)
    X = BatchNormalization(axis=3)(X)
    X = Activation('relu')(X)
    #Applying MaxPooling
    X = MaxPooling2D(pool_size= (3,3), strides= (2,2))(X)
    
    #Layer3
    X = Conv2D(384, kernel_size= (5,5), padding= 'same', kernel_initializer= glorot_uniform(seed= 0))(X)
    X = BatchNormalization(axis=3)(X)
    X = Activation('relu')(X)
    tf.nn.dropout(X,0.5)
    
    #Layer4
    X = Conv2D(384, kernel_size= (3,3), padding= 'same', kernel_initializer= glorot_uniform(seed= 0))(X)
    X = BatchNormalization(axis=3)(X)
    X = Activation('relu')(X)
    tf.nn.dropout(X,0.5)
    
    #Layer5
    X = Conv2D(256, kernel_size= (5,5), padding= 'same', kernel_initializer= glorot_uniform(seed= 0))(X)
    X = BatchNormalization(axis=3)(X)
    X = Activation('relu')(X)
    #Applying MaxPooling
    X = MaxPooling2D(pool_size= (3,3), strides= (2,2))(X)
    tf.nn.dropout(X,0.5)
    
    X = Flatten()(X)
    #Layer6
    X = Dense(1024, activation= 'relu', kernel_initializer= glorot_uniform(seed=0))(X)
    #Layer7
    tf.nn.dropout(X,0.5)
    X = Dense(256, activation= 'relu', kernel_initializer= glorot_uniform(seed=0))(X)
    #Layer8
    tf.nn.dropout(X,0.5)
    X = Dense(classes, activation= 'softmax', kernel_initializer= glorot_uniform(seed=0))(X)
    
    model = Model(X_input, X)
    
    return model
```

```python
model = AlexNet()
```

```python
model.summary()
```

```python
model.compile(optimizer = tf.keras.optimizers.Adam(learning_rate= 0.0001), loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
model.summary()
#model.fit(X_train, np.array(y_train), epochs = 40, batch_size = 128)
```

```python
# Fit the model
callbacks = [tf.keras.callbacks.EarlyStopping(patience= 5)]
history = model.fit(train_images, train_labels, validation_split= 0.2, callbacks= callbacks,
                    epochs= 100)
```
